```python
import numpy as np

def my_bisection(f, a, b, tol): 
    # approximates a root, R, of f bounded 
    # by a and b to within tolerance 
    # | f(m) | < tol with m the midpoint 
    # between a and b Recursive implementation
    
    # check if a and b bound a root
    if np.sign(f(a)) == np.sign(f(b)):
        raise Exception(
         "The scalars a and b do not bound a root")
        
    # get midpoint
    m = (a + b)/2
    
    if np.abs(f(m)) < tol:
        # stopping condition, report m as root
        return m
    elif np.sign(f(a)) == np.sign(f(m)):
        # case where m is an improvement on a. 
        # Make recursive call with a = m
        return my_bisection(f, m, b, tol)
    elif np.sign(f(b)) == np.sign(f(m)):
        # case where m is an improvement on b. 
        # Make recursive call with b = m
        return my_bisection(f, a, m, tol)
f = lambda x: x**2 - 2

r1 = my_bisection(f, 0, 2, 0.1)
print("r1 =", r1)
r01 = my_bisection(f, 0, 2, 0.01)
print("r01 =", r01)

print("f(r1) =", f(r1))
print("f(r01) =", f(r01))
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Function to solve the advection equation using upwind scheme
def upwind_solver(L, T, Nx, Nt, c):
    dx = L / Nx
    dt = T / Nt
    r = c * dt / dx

    # Initialize grid
    x = np.linspace(0, L, Nx + 1)
    t = np.linspace(0, T, Nt + 1)
    u = np.zeros((Nt + 1, Nx + 1))

    # Set initial condition
    u[0, :] = np.where((x >= 0.4) & (x <= 0.6), 1, 0)

    # Upwind time-stepping
    for n in range(0, Nt):
        for i in range(1, Nx + 1):
            u[n + 1, i] = u[n, i] - r * (u[n, i] - u[n, i - 1])

    return x, t, u

# Parameters
L = 1.0 # Length of the domain
T = 0.2 # Total simulation time
Nx = 100 # Number of spatial grid points
Nt = 200 # Number of time steps
c = 1.0 # Advection speed

# Solve the advection equation using upwind scheme
x, t, u = upwind_solver(L, T, Nx, Nt, c)

# Plot the solution
plt.figure(figsize=(8, 6))
plt.contourf(x, t, u, cmap='viridis', levels=50)
plt.colorbar(label='u')
plt.title('Upwind Scheme: Advection Equation')
plt.xlabel('Position')
plt.ylabel('Time')
plt.show()

	
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Parameters
L = 0.1 # Length of the rod in meters (10 cm)
T_initial = 0 # Initial temperature
T_left_boundary = 100 # Temperature at the left boundary
T_right_boundary = 50 # Temperature at the right boundary
t_max = 100 # Maximum simulation time in seconds
num_steps = 100 # Number of time steps
alpha = 0.01 # Thermal diffusivity

# Spatial and temporal discretization
dx = L / 100 # Spatial step size
dt = t_max / num_steps # Temporal step size

# Grid
x = np.linspace(0, L, int(L / dx) + 1)
t_values = np.linspace(0, t_max, num_steps + 1)

# Initial condition
u = np.zeros(len(x))
u[1:-1] = T_initial#initialising all the values from  the first to the second last index

# Boundary conditions
u[0] = T_left_boundary
u[-1] = T_right_boundary


def solve_heat_equation(u, alpha, dt, dx):
    k = alpha * dt / dx**2

    # Construct the tri-diagonal matrix
    A_diag = 1 + 2 * k
    A_upper = A_lower = -k
    for i in range(1, len(u) - 1):
        A_diag_i = A_diag - A_lower * A_upper / A_diag
        u[i] = (u[i] + k * (u[i-1] + u[i+1])) / A_diag_i
        A_upper = -k / A_diag_i
        
        

    return u

# Simulation
plt.figure(figsize=(10, 6))
for k_factor in [0.4, 0.6]:
    k = k_factor * 0.5
    for t in t_values:
        u[0] = T_left_boundary
        u[-1] = T_right_boundary
        u = solve_heat_equation(u, alpha, k * dt, dx)

    plt.plot(x, u, label=f'k = {k_factor}')

plt.title('Heat Conduction in a Rod')
plt.xlabel('Position (m)')
plt.ylabel('Temperature (C)')
plt.legend()
plt.show()

```

```python

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
L = 100 # Plate side length
segments = 20 # Number of segments
dx = dy = L / (segments - 1) # Grid increment
dt = 0.01 # Time step (adjust as needed)

# Create grid
x = np.linspace(0, L, segments)
y = np.linspace(0, L, segments)
X, Y = np.meshgrid(x, y)

# Initial conditions
T = np.zeros_like(X)
T[:, 0] = 0 # T(x=0) = 0
T[:, -1] = 0 # T(x=L) = 0
T[0, :] = -100 # T(y=0) = -100
T[-1, :] = 100 # T(y=L) = 100
T[0, 0] = T[0, -1] = T[-1, 0] = T[-1, -1] = (T[0, 1] + T[1, 0]) / 2 # Corner points

# Crank-Nicolson method
alpha = 0.5 * dt / dx**2
beta = 0.5 * dt / dy**2

for _ in range(1000): # Adjust the number of iterations
    Tn = T.copy()
    for i in range(1, segments - 1):
        for j in range(1, segments - 1):
            T[i, j] = alpha * (Tn[i + 1, j] - 2 * Tn[i, j] + Tn[i - 1, j]) + beta * (Tn[i, j + 1] - 2 * Tn[i, j] + Tn[i, j - 1])
            T[i, j] /= 1 + 2 * (alpha + beta)

# Surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, T, cmap='viridis')


ax.set_xlabel('Position (x)')
ax.set_ylabel('Position (y)')
ax.set_zlabel('T')
# Show plot
plt.show()




```
