```python

import numpy as np
import matplotlib.pyplot as plt


def upwind_solver(L, T, Nx, Nt, c):
    dx = L / Nx
    dt = T / Nt
    r = c * dt / dx

    
    x = np.linspace(0, L, Nx + 1)
    t = np.linspace(0, T, Nt + 1)
    u = np.zeros((Nt + 1, Nx + 1))

   
    u[0, :] = 1

    
    for n in range(0, Nt):
        for i in range(1, Nx + 1):
            u[n + 1, i] = u[n, i] - r * (u[n, i] - u[n, i - 1])

    return x, t, u


L = 1.0 
T = 0.2 
Nx = 100 
Nt = 200 
c = 1.0 


x, t, u = upwind_solver(L, T, Nx, Nt, c)


plt.figure(figsize=(8, 6))


for n in range(0, Nt + 1, int(Nt / 2)): 
    plt.plot(x, u[n, :], label=f't={t[n]:.2f}')
plt.title('Upwind Scheme:  u(x,0)=1')
plt.xlabel('Position (x)')
plt.ylabel('u(x, t)')
plt.legend()
plt.show()

	
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to solve the advection equation using central difference scheme
def central_difference_solver(L, T, Nx, Nt, c):
    dx = L / Nx
    dt = T / Nt
    r = c * dt / (2 * dx)

    # Initialize grid
    x = np.linspace(0, L, Nx + 1)
    t = np.linspace(0, T, Nt + 1)
    u = np.zeros((Nt + 1, Nx + 1))

    # Set initial condition
    u[0, :] = 1

    # Central difference time-stepping
    for n in range(0, Nt):
        for i in range(1, Nx):
            u[n + 1, i] = u[n, i] - r * (u[n, i + 1] - u[n, i - 1])

    return x, t, u

# Parameters
L = 1.0 # Length of the domain
T = 0.2 # Total simulation time
Nx = 100 # Number of spatial grid points
Nt = 100 # Number of time steps
c = 1.0 # Advection speed

# Solve the advection equation using central difference scheme
x, t, u = central_difference_solver(L, T, Nx, Nt, c)


X, T = np.meshgrid(x, t)


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, T, u, cmap='viridis', rstride=5, cstride=5, alpha=0.7)

ax.set_title('Central Difference Scheme:  with u(x,0)=1')
ax.set_xlabel('Position (x)')
ax.set_ylabel('Time (t)')
ax.set_zlabel('u(x, t)')

plt.show()

	
```

```python
import numpy as np
import matplotlib.pyplot as plt


def heat_conduction_solver(L, T, Nx, Nt, alpha):
    dx = L / Nx
    dt = T / Nt
    r = alpha * dt / dx**2

    # Initialize grid
    x = np.linspace(0, L, Nx + 1)
    t = np.linspace(0, T, Nt + 1)
    u = np.zeros((Nt + 1, Nx + 1))

    
    u[0, :]=np.sin(np.pi * x*3*0.5)
    u[:,0]=0

    
    for n in range(0, Nt):
        for i in range(1, Nx):
            u[n + 1, i] = u[n, i] + r * (u[n, i + 1] - 2 * u[n, i] + u[n, i - 1])

    return x, t, u

# Parameters
L = 1.0 
T = 0.1 
Nx = 50 
Nt = 500 
alpha = 1


x, t, u = heat_conduction_solver(L, T, Nx, Nt, alpha)
X, T = np.meshgrid(x, t)


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, T, u, cmap='viridis', rstride=5, cstride=5, alpha=0.7)

ax.set_xlabel('Position (x)')
ax.set_ylabel('Time (t)')
ax.set_zlabel('u(x, t)')

plt.show()


```
