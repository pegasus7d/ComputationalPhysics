## Euler Method


### f(x+h)= f(x) + h G(x)  G(x) is the first derivative of f(x)


```python
import numpy as np
import matplotlib.pyplot as plt

h = 0.1
x = np.arange(1, 3 + h, h)

def f(x, y):
    return x * y

s = np.zeros(len(x) + 1)  # Adjust the size for 1-based indexing
s[1] = 5  # Set the initial condition at index 1

for i in range(1, len(x)):
    s[i + 1] = s[i] + h * f(x[i - 1], s[i])

plt.plot(x, s[1:], 'bo--', label='Approximate')  # Exclude the first element for 1-based indexing
plt.plot(x, 5 * np.exp((x*x)/2 - 0.5), 'g', label='Exact')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend(loc='lower right')
plt.show()

for xx, yy in zip(x, s[1:]):
    print("x =", xx, "y =", yy)

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
v = 1.0
L = 1.0  # Length of the spatial domain
T = 1.0  # Total simulation time
dx = 0.01  # Spatial step size
dt = dx / 2.0  # Temporal step size
Nx = int(L / dx) + 1  # Number of spatial grid points
Nt = int(T / dt) + 1  # Number of temporal grid points

# Initial condition
u0 = np.ones(Nx)

# Initialize solution arrays
u_central = np.zeros((Nt, Nx))
u_upwind = np.zeros((Nt, Nx))

# Set initial condition
u_central[0, :] = u0
u_upwind[0, :] = u0

# Central Difference Scheme
for n in range(0, Nt - 1):
    for i in range(1, Nx - 1):
        u_central[n + 1, i] = u_central[n, i] - v * dt / (2 * dx) * (u_central[n, i + 1] - u_central[n, i - 1])

# Upwind Scheme
for n in range(0, Nt - 1):
    for i in range(1, Nx):
        u_upwind[n + 1, i] = u_upwind[n, i] - v * dt / dx * (u_upwind[n, i] - u_upwind[n, i - 1])

# Plotting
x_values = np.linspace(0, L, Nx)

# Plot central difference scheme
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.title('Central Difference Scheme')
plt.xlabel('x')
plt.ylabel('u(x, t)')
for n in range(0, Nt, int(Nt / 10)):
    plt.plot(x_values, u_central[n, :], label=f't = {n * dt:.2f}')
plt.legend()

# Plot upwind scheme
plt.subplot(1, 2, 2)
plt.title('Upwind Scheme')
plt.xlabel('x')
plt.ylabel('u(x, t)')
for n in range(0, Nt, int(Nt / 10)):
    plt.plot(x_values, u_upwind[n, :], label=f't = {n * dt:.2f}')
plt.legend()

plt.tight_layout()
plt.show()
```
