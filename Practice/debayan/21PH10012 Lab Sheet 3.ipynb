```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#RK2

def f(x,y):
    return (x+y)*np.sin(x)


h=0.1
x = np.arange(0, 6+h, h)
y = np.zeros(len(x))

y[0] = 5

for i in range(1, len(x)):
        k1 = h*f(x[i-1], y[i-1])
        k2 = h*f(x[i-1] + h/2, y[i-1] + k1/2)
        y[i] = y[i-1] + k2
rk2=y
plt.plot(x, y, label='Approximate (RK2)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
for xx,yy in zip(x,y):
    print("x = ",xx," y = ",yy)
```

```python

def f(x,y):
    return (x+y)*np.sin(x)

h=0.1
x = np.arange(0, 6+h, h)
y = np.zeros(len(x))

y[0] = 5

for i in range(1, len(x)):
    k1 = h * f(x[i-1], y[i-1])
    k2 = h * f(x[i-1] + h / 2, y[i-1] + k1 / 2)
    k3 = h * f(x[i-1] + h / 2, y[i-1] + k2 / 2)
    k4 = h * f(x[i-1] + h, y[i-1] + k3)
    
    y[i] = y[i-1] + (k1 + 2 * k2 + 2 * k3 + k4) / 6
rk4=y
plt.plot(x, y,  label='Approximate (RK4)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend(loc='lower right')
plt.show()

# for xx, yy in zip(x, y):
#     print("x =", xx, " y =", yy)
```

```python


def F(y, u, x):
    return (10*np.sin(x))-(5*u)-(6*y)

def exact(x):
    return -6*np.exp(-3*x)+7*np.exp(-2*x)+np.sin(x)-np.cos(x)



h=0.1
xpoints = np.arange(0, 3+h, h)
ypoints = []
upoints = []
exactpoints=[]

y = 0
u = 5
s=0;
for x in xpoints:
    ypoints.append(y)
    upoints.append(u)
    exactpoints.append(exact(x))

    m1 = h*u
    k1 = h*F(y, u, x) 
    

    m2 = h*(u + 0.5*k1)
    k2 = h*F(y+0.5*m1, u+0.5*k1, x+0.5*h)

    m3 = h*(u + 0.5*k2)
    k3 = h*F(y+0.5*m2, u+0.5*k2, x+0.5*h)

    m4 = h*(u + k3)
    k4 = h*F(y+m3, u+k3, x+h)

    y += (m1 + 2*m2 + 2*m3 + m4)/6
    u += (k1 + 2*k2 + 2*k3 + k4)/6




plt.plot(xpoints, ypoints, 'bo--',label='Approximate (RK4)')
plt.plot(xpoints, exactpoints, 'g', label='Exact')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend(loc='lower right')
plt.show()

for xx, yy in zip(xpoints, ypoints):
    print("x =", xx, " y =", yy)
```
