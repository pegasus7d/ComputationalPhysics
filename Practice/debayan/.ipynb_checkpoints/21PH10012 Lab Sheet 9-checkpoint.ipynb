```python

import numpy as np
import matplotlib.pyplot as plt

# Constants
m = 1.0 # Mass
n = 1.0 # Friction constant
T = 300.0 # Temperature (in Kelvin)
k_B = 1.38e-23 # Boltzmann constant
dt = 0.01 # Time step
total_time = 10.0 # Total simulation time

# Initial conditions
u = 1.0 # Initial position
v = 1.0 # Initial velocity


num_steps = int(total_time / dt)


positions = np.zeros(num_steps)
squared_displacements = np.zeros(num_steps)

for i in range(num_steps):
    
    random_force = np.sqrt(2 * n * k_B * T * dt / m) * np.random.normal(0, 1)

    
    k1v = -n * v + random_force
    k1u = v
    k2v = -n * (v + 0.5 * dt * k1v) + np.sqrt(2 * n * k_B * T * dt / m) * np.random.normal(0, 1)
    k2u = v + 0.5 * dt * k1v
    k3v = -n * (v + 0.5 * dt * k2v) + np.sqrt(2 * n * k_B * T * dt / m) * np.random.normal(0, 1)
    k3u = v + 0.5 * dt * k2v
    k4v = -n * (v + dt * k3v) + np.sqrt(2 * n * k_B * T * dt / m) * np.random.normal(0, 1)
    k4u = v + dt * k3v

    v = v + (dt / 6.0) * (k1v + 2 * k2v + 2 * k3v + k4v)
    u = u + (dt / 6.0) * (k1u + 2 * k2u + 2 * k3u + k4u)

    
    positions[i] = u
    squared_displacements[i] = u**2


mean_displacement = np.mean(positions)
mean_squared_displacement = np.mean(squared_displacements)

# Plot the mean squared displacement
time_points = np.arange(0, total_time, dt)
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.plot(time_points, positions, label="Position")
plt.xlabel("Time")
plt.ylabel("Position")
plt.title("Mean Displacement")
plt.grid()

plt.subplot(122)
plt.plot(time_points, squared_displacements, label="Squared Displacement")
plt.xlabel("Time")
plt.ylabel("Squared Displacement")
plt.title("Mean Squared Displacement")
plt.grid()

plt.tight_layout()
plt.show()

print(f"Mean Displacement: {mean_displacement}")
print(f"Mean Squared Displacement: {mean_squared_displacement}")

	
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Constants
num_particles = 50000
m = np.ones(num_particles) # Mass (assuming all particles have the same mass)
n = np.ones(num_particles) # Friction constant (assuming all particles have the same constant)
T = 300.0 # Temperature (in Kelvin)
k_B = 1.38e-23 # Boltzmann constant
dt = 0.01 # Time step
total_time = 10.0 # Total simulation time

# Initial conditions for all particles
u = np.ones(num_particles) # Initial positions
v = np.ones(num_particles) # Initial velocities

# Number of time steps
num_steps = int(total_time / dt)


positions = np.zeros((num_particles, num_steps))
squared_displacements = np.zeros((num_particles, num_steps))


for i in range(num_steps):
    # Generate Gaussian random forces with correct properties for all particles
    random_forces = np.sqrt(2 * n * k_B * T * dt / m) * np.random.normal(0, 1, num_particles)

    for particle in range(num_particles):
       
        k1v = -n[particle] * v[particle] + random_forces[particle]
        k1u = v[particle]
        k2v = -n[particle] * (v[particle] + 0.5 * dt * k1v) + np.sqrt(2 * n[particle] * k_B * T * dt / m[particle]) * np.random.normal(0, 1)
        k2u = v[particle] + 0.5 * dt * k1v
        k3v = -n[particle] * (v[particle] + 0.5 * dt * k2v) + np.sqrt(2 * n[particle] * k_B * T * dt / m[particle]) * np.random.normal(0, 1)
        k3u = v[particle] + 0.5 * dt * k2v
        k4v = -n[particle] * (v[particle] + dt * k3v) + np.sqrt(2 * n[particle] * k_B * T * dt / m[particle]) * np.random.normal(0, 1)
        k4u = v[particle] + dt * k3v

        
        v[particle] = v[particle] + (dt / 6.0) * (k1v + 2 * k2v + 2 * k3v + k4v)
        u[particle] = u[particle] + (dt / 6.0) * (k1u + 2 * k2u + 2 * k3u + k4u)

        # Store position and squared displacement for each particle
        positions[particle, i] = u[particle]
        squared_displacements[particle, i] = u[particle]**2

# Calculate mean and mean squared displacement for all particles
mean_displacements = np.mean(positions, axis=0)
mean_squared_displacements = np.mean(squared_displacements, axis=0)

# Plot the average mean squared displacement
time_points = np.arange(0, total_time, dt)
plt.figure(figsize=(10, 5))
plt.subplot(122)
plt.plot(time_points, mean_displacements, label="Mean Position")
plt.xlabel("Time")
plt.ylabel("Mean Position")
plt.title("Average Mean Displacement")
plt.grid()


plt.plot(time_points, mean_squared_displacements, label="Mean Squared Displacement")
plt.xlabel("Time")
plt.ylabel("Mean Squared Displacement")
plt.title("Average Mean Squared Displacement")
plt.grid()

plt.tight_layout()
plt.show()

print(f"Average Mean Displacement: {np.mean(mean_displacements)}")
print(f"Average Mean Squared Displacement: {np.mean(mean_squared_displacements)}")


```
