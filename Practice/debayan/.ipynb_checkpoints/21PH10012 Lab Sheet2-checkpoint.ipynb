```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h=0.1
x=np.arange(1,3+h,h) 
def f(x,y):
    return x*y

s=np.zeros(len(x))
s[0]=5
for i in range(0, len(x) - 1):
    s[i + 1] = s[i] + h*f(x[i], s[i])

plt.plot(x, s, 'bo--', label='Approximate')
plt.plot(x, np.exp((x*x)/2 + np.log(5) - 0.5), 'g', label='Exact')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
for xx,yy in zip(x,s):
    print("x = ",xx," y = ",yy)
```

```python
h=0.1
x = np.arange(1, 3+h, h)
y = np.zeros(len(x))

y[0] = 5

for i in range(1, len(x)):
        k1 = h*f(x[i-1], y[i-1])
        k2 = h*f(x[i-1] + h/2, y[i-1] + k1/2)
        y[i] = y[i-1] + k2

plt.plot(x, y, 'bo--', label='Approximate')
plt.plot(x, np.exp((x*x)/2 + np.log(5) - 0.5), 'g', label='Exact')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
for xx,yy in zip(x,y):
    print("x = ",xx," y = ",yy)
```
