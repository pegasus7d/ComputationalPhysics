```python
import numpy as np
import matplotlib.pyplot as plt

# Define the BVP equation
def bvp_equation(x, y, yp):
    return -y
def f(x):
    return -3 * np.sin(x)
# Define the RK4 method for solving IVP
def solve_ivp_rk4(bvp_func, x0, y0, yp0, x_final, num_points):
    h = (x_final - x0) / num_points
    x_values = [x0]
    y_values = [y0]
    yp_values = [yp0]

    for _ in range(num_points):
        x = x_values[-1]
        y = y_values[-1]
        yp = yp_values[-1]

        k1 = h * yp
        k2 = h * bvp_func(x + h/2, y + k1/2, yp + k1/2)
        k3 = h * bvp_func(x + h/2, y + k2/2, yp + k2/2)
        k4 = h * bvp_func(x + h, y + k3, yp + k3)

        y_next = y + (k1 + 2*k2 + 2*k3 + k4) / 6
        yp_next = yp + (k1 + 2*k2 + 2*k3 + k4) / 6

        x_values.append(x + h)
        y_values.append(y_next)
        yp_values.append(yp_next)

    return x_values, y_values

# Define the target boundary conditions
target_y_a = 0
target_y_b = -3

# Set the initial range for guessing y'(0)
yp0_min = -20
yp0_max = 20
tolerance = 1e-5

# Perform shooting method with a binary search
while yp0_max - yp0_min > tolerance:
    
    yp0_guess = (yp0_min + yp0_max) / 2

    
    x0 = 0

    # Determine initial conditions for the IVP
    y0 = 0
    yp0 = yp0_guess

    # Solve the IVP using RK4 method
    x_values, y_values = solve_ivp_rk4(bvp_equation, x0, y0, yp0, 1, 1000)

    
    if y_values[-1] > target_y_b:
        yp0_max = yp0_guess
    else:
        yp0_min = yp0_guess

# Final solution
final_x_values, final_y_values = solve_ivp_rk4(bvp_equation, x0, y0, yp0_min, 1, 1000)
y_analytical = f(np.array(x_values))
# Plot the numerical solution
plt.plot(final_x_values, final_y_values, label='Numerical Solution')
plt.plot(final_x_values, y_analytical, label='Analytical Solution')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Shooting Method with RK4 for BVP')
plt.grid(True)
plt.legend()
plt.show()

print(f"Final y(1) = {final_y_values[-1]}")
print(f"Final y'(0) = {yp0_min}")

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the BVP equation
def bvp_equation(x, y, yp):
    return 2*y+8*x*(9-x)

# Define the RK2 method for solving IVP
def solve_ivp_rk2(bvp_func, x0, y0, yp0, x_final, num_points):
    h = (x_final - x0) / num_points
    x_values = [x0]
    y_values = [y0]
    yp_values = [yp0]

    for _ in range(num_points):
        x = x_values[-1]
        y = y_values[-1]
        yp = yp_values[-1]

        k1 = h * yp
        k2 = h * bvp_func(x + h, y + k1, yp + k1)

        y_next = y + (k1 + k2) / 2
        yp_next = yp + (k1 + k2) / 2

        x_values.append(x + h)
        y_values.append(y_next)
        yp_values.append(yp_next)

    return x_values, y_values

# Define the target boundary conditions
target_y_a = 0
target_y_b = 0

# Set the initial range for guessing y'(0)
yp0_min = -20
yp0_max = 20
tolerance = 1e-5

# Perform shooting method with a binary search
while yp0_max - yp0_min > tolerance:
    # Choose a new guess for y'(0) in the middle of the current range
    yp0_guess = (yp0_min + yp0_max) / 2

    
    x0 = 0

    # Determine initial conditions for the IVP
    y0 = 0
    yp0 = yp0_guess

    # Solve the IVP using RK2 method
    x_values, y_values = solve_ivp_rk2(bvp_equation, x0, y0, yp0, 1, 1000)

    # Check the boundary condition at x=1 and adjust the range accordingly
    if y_values[-1] > target_y_b:
        yp0_max = yp0_guess
    else:
        yp0_min = yp0_guess

# Final solution
final_x_values, final_y_values = solve_ivp_rk2(bvp_equation, x0, y0, yp0_min, 1, 1000)

# Plot the numerical solution
plt.plot(final_x_values, final_y_values, label='Numerical Solution')

plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Shooting Method with RK2 for BVP')
plt.grid(True)
plt.legend()
plt.show()

print(f"Final y(1) = {final_y_values[-1]}")
print(f"Final y'(0) = {yp0_min}")
```
