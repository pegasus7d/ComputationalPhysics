```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#euler
h=0.1
x=np.arange(1,3+h,h) 
def f(x,y):
    return x*y

s=np.zeros(len(x))
s[0]=5
for i in range(0, len(x) - 1):
    s[i + 1] = s[i] + h*f(x[i], s[i])

plt.plot(x, s, 'bo--', label='Approximate')
plt.plot(x, np.exp((x*x)/2 + np.log(5) - 0.5), 'g', label='Exact')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
for xx,yy in zip(x,s):
    print("x = ",xx," y = ",yy)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x * y

h = 0.1
x = np.arange(1, 3 + h, h)
y = np.zeros(len(x))

y[0] = 5

for i in range(1, len(x)):
    k1 = h * f(x[i-1], y[i-1])
    k2 = h * f(x[i-1] + h / 2, y[i-1] + k1 / 2)
    k3 = h * f(x[i-1] + h / 2, y[i-1] + k2 / 2)
    k4 = h * f(x[i-1] + h, y[i-1] + k3)
    
    y[i] = y[i-1] + (k1 + 2 * k2 + 2 * k3 + k4) / 6

plt.plot(x, y, 'bo--', label='Approximate (RK4)')
plt.plot(x, np.exp((x * x) / 2 + np.log(5) - 0.5), 'g', label='Exact')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend(loc='lower right')
plt.show()

for xx, yy in zip(x, y):
    print("x =", xx, " y =", yy)
```

```python
import numpy as np

def my_bisection(f, a, b, tol): 
    # approximates a root, R, of f bounded 
    # by a and b to within tolerance 
    # | f(m) | < tol with m the midpoint 
    # between a and b Recursive implementation
    
    # check if a and b bound a root
    if np.sign(f(a)) == np.sign(f(b)):
        raise Exception(
         "The scalars a and b do not bound a root")
        
    # get midpoint
    m = (a + b)/2
    
    if np.abs(f(m)) < tol:
        # stopping condition, report m as root
        return m
    elif np.sign(f(a)) == np.sign(f(m)):
        # case where m is an improvement on a. 
        # Make recursive call with a = m
        return my_bisection(f, m, b, tol)
    elif np.sign(f(b)) == np.sign(f(m)):
        # case where m is an improvement on b. 
        # Make recursive call with b = m
        return my_bisection(f, a, m, tol)
f = lambda x: x**2 - 2

r1 = my_bisection(f, 0, 2, 0.1)
print("r1 =", r1)
r01 = my_bisection(f, 0, 2, 0.01)
print("r01 =", r01)

print("f(r1) =", f(r1))
print("f(r01) =", f(r01))
```

```python
import numpy as np

f = lambda x: x**2 - 2
f_prime = lambda x: 2*x
newton_raphson = 1.4 - (f(1.4))/(f_prime(1.4))

print("newton_raphson =", newton_raphson)
print("sqrt(2) =", np.sqrt(2))
def my_newton(f, df, x0, tol):
    # output is an estimation of the root of f 
    # using the Newton Raphson method
    # recursive implementation
    if abs(f(x0)) < tol:
        return x0
    else:
        return my_newton(f, df, x0 - f(x0)/df(x0), tol)
```

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: x**2 - 2
f_prime = lambda x: 2*x

x0 = 1.4
tol = 1e-6

x_values = [x0]
while abs(f(x0)) >= tol:
    x0 = x0 - f(x0) / f_prime(x0)
    x_values.append(x0)

root_approximation = x_values[-1]

print("Approximated root using Newton-Raphson:", root_approximation)
print("Exact square root of 2:", np.sqrt(2))

# Plotting the convergence graph
plt.plot(x_values, 'bo-')
plt.axhline(y=np.sqrt(2), color='r', linestyle='--', label='Exact sqrt(2)')
plt.xlabel('Iteration')
plt.ylabel('Approximated Root')
plt.title('Convergence of Newton-Raphson Method')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def my_bisection(f, a, b, tol): 
    # Recursive implementation of the bisection method
    
    # check if a and b bound a root
    if np.sign(f(a)) == np.sign(f(b)):
        raise Exception("The scalars a and b do not bound a root")
        
    # get midpoint
    m = (a + b) / 2
    
    if np.abs(f(m)) < tol:
        # stopping condition, report m as root
        return m
    elif np.sign(f(a)) == np.sign(f(m)):
        # case where m is an improvement on a. 
        # Make recursive call with a = m
        return my_bisection(f, m, b, tol)
    elif np.sign(f(b)) == np.sign(f(m)):
        # case where m is an improvement on b. 
        # Make recursive call with b = m
        return my_bisection(f, a, m, tol)

# Define the function and its bounds
f = lambda x: x**2 - 2
a = 0
b = 2

# Perform bisection
tolerances = [0.1, 0.01]
roots = []

for tol in tolerances:
    root = my_bisection(f, a, b, tol)
    roots.append(root)

# Plot the graph of the function and estimated roots
x_vals = np.linspace(a, b, 400)
y_vals = f(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='$f(x) = x^2 - 2$')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')

for root, tol in zip(roots, tolerances):
    plt.plot(root, f(root), 'ro', label=f'Root (Tolerance {tol})')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of the Function with Estimated Roots')
plt.legend()
plt.grid()
plt.show()
```
