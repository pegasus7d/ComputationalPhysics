```python
# plt.figure(figsize=(8,3))
plt.plot(x,y,'o--',color='purple',lw=4,label='Component 1')
plt.ylabel('Voltage [V]')
plt.xlabel('Time[s]')
plt.legend(loc='upper right')
```

```python
x2=np.linspace(0,15,100)
y2=np.sin(x2)
```
